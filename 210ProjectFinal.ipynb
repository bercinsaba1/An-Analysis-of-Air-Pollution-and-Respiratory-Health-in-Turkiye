# Hypothesis Testing

Statistical Testing of the Relationship between PM10 pollution and deaths year by year 
- Null Hypothesis: There is not significant association between PM10 pollution and deaths caused by respiratory health issues.
- Alternative Hypothesis: There is significant association between PM10 pollution and deaths caused by respiratory health issues.

```python
from scipy.stats import ttest_ind, chi2_contingency
from scipy import stats
```

```python
alpha = 0.05
random_state = 35 
```

Three years' p values are examined from chi-square test.

```python
crosstab = pd.crosstab(filled_df['pollution2010'], ratios_df['2010'])

chi2, p, dof, expected = stats.chi2_contingency(crosstab)

crosstab2 = pd.crosstab(filled_df['pollution2013'], ratios_df['2013'])

chi2, p2, dof, expected = stats.chi2_contingency(crosstab2)

crosstab3 = pd.crosstab(filled_df['pollution2017'], ratios_df['2017'])

chi2, p3, dof, expected = stats.chi2_contingency(crosstab3)

print("p-value for 2013", p2)
print("p-value for 2010", p)
print("p-value for 2017", p3)
```
